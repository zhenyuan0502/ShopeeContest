```python
def action_1(is_left):
    return 0, is_left

def action_2(line, m, is_left):
    if is_left:
        return sum(line[:m]), is_left
    
    return sum(line[-m:]), is_left

def get_max_consecutive(line):
    prefix_sum = [line[0]]

    for num in line[1:]:
        prefix_sum.append(prefix_sum[-1] + num)

    return max(prefix_sum)

def action_2_max(line, is_left):
    if is_left:
        return get_max_consecutive(line), is_left
    
    return get_max_consecutive(list(reversed(line))), is_left

def action_3(line, is_left):
    return sum(line), not(is_left)
```

```python
line = [1, 2, 3, 4]
action_2(line, 3, False)
action_1(True)
# action_3(line, False)
```

```python
# N dòng => (Act1, Act1,..., N), (Act1, Act2,..., N),....
def get_combination(N):
    result = []
    ranges = N*[3]
    size = 3**N
    lenr = len(ranges)

    for i in range(size):
        perm = []
        d, r = i, 0
        for rng_i in (1, lenr):
            d, r = divmod(d, ranges[lenr - rng_i])
            perm.append(r)
        perm.extend([0]*(lenr-len(perm)))   # pad with zeros
        result.append(list(reversed(perm)))  

    return result

get_combination(10)
```

```python
def exec(matrix):
    totals = []
    for actions in get_combination(len(matrix)):
        # print(actions)
        is_left = True
        total = 0
        for day in range(len(matrix)):
            if actions[day] == 0:
                # print('Call Act 1 for ', matrix[day])
                _, is_left = action_1(is_left)
                continue

            if actions[day] == 1:
                # print('Call Act 2 for ', matrix[day])
                sub_sum, is_left = action_2_max(matrix[day], is_left)
                total += sub_sum
                continue

            if actions[day] == 2:
                # print('Call Act 3 for ', matrix[day])
                sub_sum, is_left = action_3(matrix[day], is_left)
                total += sub_sum
                continue
        totals.append(total)

    return max(totals)

```

```python
matrix = [[1, 4, -5], 
          [-1, -9, 100], 
          [-4999, 9, 100]]

exec(matrix)
```
