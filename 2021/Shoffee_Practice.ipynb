```python
def shoffee(N, K, Vs):
    count = 0
    for i in range(len(Vs)):
        for j in range(i, len(Vs)):
            # print(f"{i} {j} {Vs[i:j+1]} {sum(Vs[i:j+1])/(j-i+1)}")
            if sum(Vs[i:j+1])/(j-i+1) >= K:
                count += 1

    return count

def opt_shoffee(N, K, Vs):
    count = 0
    for i in range(len(Vs)):
        sum_sliding = 0
        sum_size = 0
        for j in range(i, len(Vs)):
            sum_sliding += Vs[j]
            sum_size += 1
            # print(f"{i} {j} ({sum_sliding}) {sum_sliding/(j-i+1)}")
            if sum_sliding/(sum_size) >= K:
                # print(f"{i} {j} {Vs[i: j + 1]} = {sum_sliding}  => {sum_sliding/(j-i+1)}")
                count += 1
        
    return count

import bisect 

def search_greater_than(arr, n, l = 0, r = None):
    if arr == []:
        return []

    if len(arr) == 1:
        return [] if arr[0] < n else [arr[0]]

    if r is None:
        r = len(arr)

    mid = (r + l) // 2
    if r <= l:
        if arr[mid] < n:
            return arr[mid+1:]
        return arr[mid:]

    if arr[mid] < n:
        print('right', mid + 1, mid, l, r)
        return search_greater_than(arr, n, mid+1, r)
    elif n < arr[mid]:
        print('left', mid - 1, mid, l, r)
        return search_greater_than(arr, n, l, mid-1)
    else:
        if arr[mid] < n:
            return arr[mid+1:]
        return arr[mid:]


def opt_ps_shoffee(N, K, Vs):
    count = 0
    prefixSum = 0
    mp = {}
    storedKey = []
    for i in range(len(Vs)):
        prefixSum += Vs[i] - K
        if prefixSum >= 0:
            count += 1

        end = bisect.bisect_right(storedKey, prefixSum)
        # print(prefixSum, storedKey, storedKey[:end], mp)
        for key in storedKey[:end]:
            count += mp[key]
        
        # print(len(mp))
        # can optimize here
        # for key in mp:
        #     if prefixSum - key >= 0:
        #         count += mp[key]

        if prefixSum in mp:  
            mp[prefixSum] += 1
        else:
            mp[prefixSum] = 1
            bisect.insort(storedKey, prefixSum)

        # print(f"{Vs[i]}-{K} = {Vs[i]-K},  prefixSum={prefixSum}  ,{mp}, count = {count}")
    
    return count

def opt_ps_2_shoffee(N, K, Vs):
    count = 0
    prefixSum = 0
    mp = {}
    storedKey = []
    for i in range(len(Vs)):
        prefixSum += Vs[i] - K
        if prefixSum >= 0:
            count += 1

        end = bisect.bisect_right(storedKey, prefixSum)
        # print(prefixSum, storedKey, storedKey[:end], mp)
        for key in storedKey[:end]:
            count += mp[key]
        
        # print(len(mp))
        # can optimize here
        # for key in mp:
        #     if prefixSum - key >= 0:
        #         count += mp[key]

        if prefixSum in mp:  
            mp[prefixSum] += 1
        else:
            mp[prefixSum] = 1
            bisect.insort(storedKey, prefixSum)

        # print(f"{Vs[i]}-{K} = {Vs[i]-K},  prefixSum={prefixSum}  ,{mp}, count = {count}")
    
    return count

# shoffee(6, 3, [1, 1, 4, 5, 1, 4])
# opt_shoffee(6, 3, [1, 3, 4, 2, 3, 3])
# opt_shoffee(4, 3, [7, 6, 2, 1, 3, 10])
# opt_ps_shoffee(4, 3, [7, 6, 2, 1, 3, 10])

# shoffee(4, 2, [2, 1, 3])
opt_ps_shoffee(4, 2, [2, 1, 3, 6])
opt_shoffee(4, 2, [2, 1, 3, 6])

# opt_ps_shoffee(4, 3, [1, 3, 4, 2])
# opt_ps_shoffee(4, 3, [1, 3, 4, 2, 3, 3])

# opt_ps_shoffee(10, 6, [12, 5, 3, 10, 4, 8, 10, 12, -6, -1])

# opt_shoffee(3, 3, [1, 3, 4])

# search_greater_than([1, 3, 4, 7, 9, 11, 15], 2.9)
# search_greater_than([2], 1.8)

# bisect.bisect_left([1, 3, 4, 7, 9, 11, 15], 9.5)
# bisect.bisect_left([3], 1)
```
